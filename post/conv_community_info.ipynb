```python
import pandas as pd
import re
from uuid import uuid4
```

```python
community = pd.read_json("../community_info.jsonl", lines=True)
community.info()
```

```python
community_info = pd.DataFrame(community['info'].tolist())
community = pd.concat([community, community_info], axis=1)
community.info()
```

```python
community['uuid'] = [uuid4() for _ in range(community.shape[0])]
community.info()
```

```python
com_add = community[['uuid', 'name']]
com_add.loc[pd.notna(community['楼盘地址']), 'address'] = community['楼盘地址']
com_add.loc[pd.notna(community['小区地址']), 'address'] = community['小区地址']
com_add.head()
```

```python
com_add.loc[:, 'address'] = [re.subn(r'地图$', '', x)[0].strip() for x in com_add['address']]
```

```python
com_add.to_csv('../community.csv')
```
